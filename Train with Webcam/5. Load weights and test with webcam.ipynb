```python
import os
import cv2
%matplotlib inline
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
#   ( = model.train    save_dir   )
model = YOLO("C:/Users/USER/runs/detect/train11/weights/best.pt")
#  
cap = cv2.VideoCapture(0)
#   
if not cap.isOpened():
    print("camera open failed")
    exit()
while True:
    #  
    status, img = cap.read()
    if not status:
        print("Can't read camera")
        break
    #   
    results = model.predict(source = img)
    plots = results[0].plot()
    cv2.imshow("PC_Webcam",plots)
    # esc  
    if cv2.waitKey(1) == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
